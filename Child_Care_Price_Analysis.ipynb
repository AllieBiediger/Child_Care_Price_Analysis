```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("NDCP2022.csv")
```

```python
plt.figure(figsize=(12, 6))
```

```python
median_price_b5 = df.groupby('STUDYYEAR')['MCBto5'].median()
median_price_6to11 = df.groupby('STUDYYEAR')['MC6to11'].median()

# Median Household Income
median_income = df.groupby('STUDYYEAR')['MHI'].median()
```

```python
state_income = df.groupby('STATE_NAME')['MHI'].mean().sort_values(ascending=False)

# Plot State Income
plt.figure(figsize=(12, 8))
state_income.plot(kind="bar", color=plt.cm.rainbow(np.linspace(0, 1, len(state_income))))

plt.title("Average Median Household Income by State")
plt.xlabel("State")
plt.ylabel("Average Income ($)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
state_price = df.groupby('STATE_NAME')['MCBto5'].mean()

# Calculate price-to-income ratio
price_to_income_ratio = (state_price / state_income).sort_values()

# Plot Price-to-Income Ratio
plt.figure(figsize=(12, 8))
price_to_income_ratio.plot(kind="bar", color=plt.cm.viridis(np.linspace(0, 1, len(price_to_income_ratio))))

plt.title("Price-to-Income Ratio by State")
plt.xlabel("State")
plt.ylabel("Ratio")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
best_ratio_state = price_to_income_ratio.idxmin()
best_ratio_value = price_to_income_ratio.min()
print(f"The state with the best price-to-income ratio is {best_ratio_state} with a ratio of {best_ratio_value:.4f}.")
```

```python
worst_state = price_to_income_ratio.idxmax()
worst_ratio = price_to_income_ratio.max()

print(f"The state with the worst price-to-income ratio is {worst_state} with a ratio of {worst_ratio:.4f}.")
```

```python
plt.figure(figsize=(12, 6))

# Yearly Average Median Prices for MCBto5
df.groupby('STUDYYEAR')['MCBto5'].mean().plot(label="Average Price (MCBto5)", marker="o")

plt.title("Average Price for MCBto5 Over Years")
plt.xlabel("Year")
plt.ylabel("Price ($)")
plt.legend()
plt.grid()
plt.show()
```

```python
average_income_over_time = df.groupby('STUDYYEAR')['MHI'].mean()

# Plot the trend of income over time
plt.figure(figsize=(12, 6))
plt.plot(average_income_over_time, marker='o', linestyle='-', color='teal', label="Average Median Household Income")

plt.title("Average Median Household Income Over Time", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Income ($)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Filter the data for Washington State
wa_income_data = df[df['STATE_NAME'] == 'Washington']

# Group by year and calculate the average income
wa_income_trend = wa_income_data.groupby('STUDYYEAR')['MHI'].mean()

# Plot the income trend over time
plt.figure(figsize=(10, 6))
plt.plot(wa_income_trend.index, wa_income_trend.values, marker='o', linestyle='-', color='blue')
plt.title('Median Household Income in Washington State Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Household Income ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(wa_income_trend.index, rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Filter the data for Washington and South Dakota
wa_income_data = df[df['STATE_NAME'] == 'Washington']
sd_income_data = df[df['STATE_NAME'] == 'South Dakota']

# Group by year and calculate the average income
wa_income_trend = wa_income_data.groupby('STUDYYEAR')['MHI'].mean()
sd_income_trend = sd_income_data.groupby('STUDYYEAR')['MHI'].mean()

# Plot for Washington State
plt.figure(figsize=(10, 6))
plt.plot(wa_income_trend.index, wa_income_trend.values, marker='o', linestyle='-', color='blue', label='Washington')
plt.title('Median Household Income in Washington State Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Household Income ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(wa_income_trend.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Plot for South Dakota
plt.figure(figsize=(10, 6))
plt.plot(sd_income_trend.index, sd_income_trend.values, marker='o', linestyle='-', color='green', label='South Dakota')
plt.title('Median Household Income in South Dakota Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Household Income ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(sd_income_trend.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Filter the data for Washington and South Dakota
wa_income_data = df[df['STATE_NAME'] == 'Washington']
sd_income_data = df[df['STATE_NAME'] == 'South Dakota']

# Group by year and calculate the average income
wa_income_trend = wa_income_data.groupby('STUDYYEAR')['MHI'].mean()
sd_income_trend = sd_income_data.groupby('STUDYYEAR')['MHI'].mean()

# Plotting both trends on the same graph
plt.figure(figsize=(12, 6))
plt.plot(wa_income_trend.index, wa_income_trend.values, marker='o', linestyle='-', color='blue', label='Washington')
plt.plot(sd_income_trend.index, sd_income_trend.values, marker='o', linestyle='--', color='green', label='South Dakota')

# Adding titles and labels
plt.title('Median Household Income Trends: Washington vs. South Dakota', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Household Income ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(wa_income_trend.index, rotation=45)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Filter the data for Washington and South Dakota
wa_bto5_data = df[df['STATE_NAME'] == 'Washington']
sd_bto5_data = df[df['STATE_NAME'] == 'South Dakota']

# Group by year and calculate the average Bto5 price
wa_bto5_trend = wa_bto5_data.groupby('STUDYYEAR')['MCBto5'].mean()
sd_bto5_trend = sd_bto5_data.groupby('STUDYYEAR')['MCBto5'].mean()

# Plotting both trends on the same graph
plt.figure(figsize=(12, 6))
plt.plot(wa_bto5_trend.index, wa_bto5_trend.values, marker='o', linestyle='-', color='red', label='Washington')
plt.plot(sd_bto5_trend.index, sd_bto5_trend.values, marker='o', linestyle='--', color='purple', label='South Dakota')

# Adding titles and labels
plt.title('Average Bto5 Price: Washington vs. South Dakota', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Bto5 Price ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(wa_bto5_trend.index, rotation=45)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Calculate the overall average Bto5 price for both states
avg_bto5_price = df[df['STATE_NAME'].isin(['Washington', 'South Dakota'])].groupby('STATE_NAME')['MCBto5'].mean()

# Bar chart
plt.figure(figsize=(8, 6))
avg_bto5_price.plot(kind='bar', color=['red', 'purple'], alpha=0.7)
plt.title('Average Bto5 Price: Washington vs. South Dakota', fontsize=16)
plt.ylabel('Average Bto5 Price ($)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Calculate average income and Bto5 price for both states
state_data = df[df['STATE_NAME'].isin(['Washington', 'South Dakota'])].groupby('STATE_NAME').agg({
    'MHI': 'mean',
    'MCBto5': 'mean'
})

# Stacked bar chart
state_data.plot(kind='bar', stacked=True, figsize=(8, 6), color=['blue', 'orange'], alpha=0.8)
plt.title('Income and Bto5 Price Comparison: Washington vs. South Dakota', fontsize=16)
plt.ylabel('Amount ($)', fontsize=12)
plt.xticks(rotation=0)
plt.legend(['Median Household Income (MHI)', 'Bto5 Price'])
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Scatterplot data for the two states
wa_data = df[df['STATE_NAME'] == 'Washington']
sd_data = df[df['STATE_NAME'] == 'South Dakota']

plt.figure(figsize=(10, 6))
plt.scatter(wa_data['MHI'], wa_data['MCBto5'], color='red', alpha=0.7, label='Washington')
plt.scatter(sd_data['MHI'], sd_data['MCBto5'], color='purple', alpha=0.7, label='South Dakota')

# Adding labels and titles
plt.title('Price vs. Income: Washington vs. South Dakota', fontsize=16)
plt.xlabel('Median Household Income (MHI)', fontsize=12)
plt.ylabel('Bto5 Price ($)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.6)
plt.tight_layout()
plt.show()
```

```python

# Filter the data for Washington and South Dakota
washington_data = df[df['STATE_NAME'] == 'Washington']
south_dakota_data = df[df['STATE_NAME'] == 'South Dakota']

# Group by 'STUDYYEAR' and calculate the mean MHI for both states
washington_income = washington_data.groupby('STUDYYEAR')['MHI'].mean()
south_dakota_income = south_dakota_data.groupby('STUDYYEAR')['MHI'].mean()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(washington_income.index, washington_income, label='Washington', color='blue', marker='o')
plt.plot(south_dakota_income.index, south_dakota_income, label='South Dakota', color='green', marker='o')

plt.title('Median Household Income Over Time (Washington vs South Dakota)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Household Income ($)', fontsize=12)
plt.legend()

# Show the plot
plt.show()
```

```python
# Group by year and calculate averages for employment rates
washington_employment = washington_data.groupby('STUDYYEAR')[['MEMR_20to64', 'FEMR_20to64']].mean()
south_dakota_employment = south_dakota_data.groupby('STUDYYEAR')[['MEMR_20to64', 'FEMR_20to64']].mean()

# Plot
plt.figure(figsize=(14, 7))

# Washington
plt.plot(washington_employment.index, washington_employment['MEMR_20to64'], label='Male (WA)', color='blue', linestyle='--')
plt.plot(washington_employment.index, washington_employment['FEMR_20to64'], label='Female (WA)', color='blue', linestyle='-')

# South Dakota
plt.plot(south_dakota_employment.index, south_dakota_employment['MEMR_20to64'], label='Male (SD)', color='green', linestyle='--')
plt.plot(south_dakota_employment.index, south_dakota_employment['FEMR_20to64'], label='Female (SD)', color='green', linestyle='-')

plt.title('Employment Rates (20-64 Years) Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Employment Rate (%)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Group by year and sum up total population for Washington and South Dakota
washington_population = washington_data.groupby('STUDYYEAR')['TOTALPOP'].sum()
south_dakota_population = south_dakota_data.groupby('STUDYYEAR')['TOTALPOP'].sum()

# Create a DataFrame for easier comparison
population_comparison = pd.DataFrame({
    'Washington': washington_population,
    'South Dakota': south_dakota_population
})

# Plot a bar chart
population_comparison.plot(kind='bar', figsize=(12, 6), color=['blue', 'green'], alpha=0.8)

plt.title('Total Population Comparison: Washington vs South Dakota', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.legend(title='State')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
